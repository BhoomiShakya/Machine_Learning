## SMOTE (Synthetic Minority Oversampling Technique) Summary

**What is SMOTE?**
- SMOTE is a technique used to handle imbalanced datasets by generating synthetic samples for the minority class
- It works by creating new synthetic examples between existing minority class samples
- This helps improve model performance on imbalanced datasets

**Key Benefits:**
1. Reduces overfitting compared to simple oversampling
2. Creates more realistic synthetic samples
3. Improves model performance on minority class
4. Maintains the original data distribution characteristics

**When to Use SMOTE:**
- When you have a severely imbalanced dataset
- When simple oversampling leads to overfitting
- When you need to improve recall for minority class
