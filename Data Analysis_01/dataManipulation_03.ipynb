Handling Missing Values


```python
print(df.isnull())
```

```python
#filling the missing values with 0
print(df.fillna(0)) 
```

```python
#fillling the missing values with mean of the column
df['Odometer_fillNA']= df['Odometer (KM)'].fillna(df['Odometer (KM)'].mean)
df[0:51]
```

```python
#Renaming Columns
df = df.rename(columns={'Make':'Company'})
df
```

```python
#change datatypes
df['Door_num']=df['Doors'].fillna(df['Doors'].mean()).astype(int)
df.iloc[0:20, :]
```

```python
df['Door_num'] = df['Doors'].apply(lambda x: x*2)
df
```

```python
print(df['Doors'].sum())
grouped_sum= df.groupby('Company')['Price'].sum()
print(grouped_sum)
```

```python
grouped_agg= df.groupby('Company')['Price'].agg(['mean','sum','count'])
print(grouped_agg)
```

Merging & joining DataFrame

```python
data2=[
   { 'A':'a', 'B' :'b', 'C':'c'},
    ]

data3=[
   { 'A':'x', 'B' :'y', 'C':'z'},
    ]

df2=pd.DataFrame(data2)
df3=pd.DataFrame(data3)

print(df2)
print(df3)
```

```python
pd.merge(df2,df3)
```
